```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
a = 11 + 1.5*np.random.randn(60)
a = np.append(a,11 + 1.7*np.random.randn(60))
a_tag = np.full(60, 'B2C')
a_tag = np.append(a_tag,np.full(60, 'C2B'))
a_source = np.full(30, '汽车之家')
b_source = np.full(30, '易车')
a_source = np.append(a_source, b_source,)
a_source = np.append(a_source, a_source,)
```

```python
df = pd.DataFrame({'价格':a, '场景':a_tag, '来源': a_source})
```

# Violinplot

```python
fig = {
    "data": [
        {
            "type": 'violin',
            "x": df[df['场景'] == 'B2C']['来源'],
            "y": df[df['场景'] == 'B2C']['价格'],
            "legendgroup": 'B2C',
            "scalegroup": 'B2C',
            "name": 'B2C',
            "side": 'negative',
            "box": {
                "visible": True
            },
            "meanline": {
                "visible": True
            },
            "line": {
                "color": 'steelblue'
            }
        },
        {
            "type": 'violin',
            "x": df[df['场景'] == 'C2B']['来源'],
            "y": df[df['场景'] == 'C2B']['价格'],
            "legendgroup": 'C2B',
            "scalegroup": 'C2B',
            "name": 'C2B',
            "side": 'positive',
            "box": {
                "visible": True
            },
            "meanline": {
                "visible": True
            },
            "line": {
                "color": 'coral'
            }
        }
    ],
    "layout" : {
        "yaxis": {
            "title":'价格(万元)',
            "zeroline": False,
        },
        "title":'不同场景的报价分布',
        #"violingap": 10,
        "violinmode": "overlay"
    }
}


iplot(fig, filename = 'violin', validate = False)
```
